```python
import duckdb
import pandas as pd

%load_ext sql
```

```python
bigmac = pd.read_csv("data/bigmac.csv")
```

```python
query = """
SELECT 
    CAST(date AS date),
    name,
    dollar_price
FROM 
    bigmac
WHERE 
    date >= '2023-01-01'
ORDER BY 
    dollar_price DESC
;
"""

duckdb.sql(query)
```

```python
(
    bigmac
    .loc[lambda df_: df_['date'] >= "2023-01-01", ["name", "dollar_price"]]
    .sort_values("dollar_price", ascending=False)
)
```
